## Reading Text Files in Pieces

```python
result = pd.read_csv("examples/ex6.csv")
result
```

```python
# you can only read a few rows
pd.read_csv("examples/ex6.csv", nrows=5)
```

```python
# or read the file in chunks
chunker = pd.read_csv("examples/ex6.csv", chunksize=1000)
type(chunker)
```

```python
# you can now iterate the chunks
chunker = pd.read_csv("examples/ex6.csv", chunksize=1000)
tot = pd.Series([], dtype="int64")
for piece in chunker:
    tot = tot.add(piece["key"].value_counts(), fill_value=0)
tot = tot.sort_values(ascending=False)
tot
```

## Writing Data to Text Format

```python
data = pd.read_csv("examples/ex5.csv")
data
```

```python
# simply using to_csv
data.to_csv("examples/out.csv")
!cat examples/out.csv
```

```python
import sys

# can also use other delimiters
data.to_csv(sys.stdout, sep="|")
```

```python
# missing values can be handled explicitly
data.to_csv(sys.stdout, na_rep="NULL")
```

```python
# index and header can be handled explicitly as well
data.to_csv(sys.stdout, index=False, header=False)
```

```python
data.to_csv(sys.stdout, index=False, columns=["a", "b", "c"])
```

# Interacting with Databases

```python
import sqlite3

query = """
CREATE TABLE test (a VARCHAR(20), b VARCHAR(20), c REAL, d INTEGER)
"""
con = sqlite3.connect("mydata.sqlite")
con.execute(query)
con.commit()
```

```python
data = [
    ("Atlanta", "Georgia", 1.25, 6),
    ("Tallahassee", "Florida", 2.6, 3),
    ("Sacramento", "California", 1.7, 5)
]
stmt = "INSERT INTO test VALUES(?, ?, ?, ?)"
con.executemany(stmt, data)
con.commit()
```

```python
cursor = con.execute("SELECT * FROM test")
rows = cursor.fetchall()
rows
```

```python
pd.DataFrame(rows, columns=[x[0] for x in cursor.description])
```
